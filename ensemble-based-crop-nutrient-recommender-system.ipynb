  üìÇ Dataset Overview
</h3>

This project uses the **Fertilizer Prediction** dataset ([Kaggle Playground Series ‚Äì S5E6, 2025](https://www.kaggle.com/competitions/playground-series-s5e6/overview)) to recommend optimal fertilizers based on climate, soil, and crop features. Main feature include temperature, humidity, moisture, soil type, crop type, and soil nutrients (N, P, K). The target is the recommended fertilizer name.  

### Provided files are 

* `train.csv` (750,000 samples): Contains features and the target fertilizer label.
* `test.csv` (250,000 samples): Requires predicting up to **three** fertilizer types per row.
* `Fertilizer Prediction.csv` (100,000 samples): The original dataset used to generate the synthetic competition data (sources: [IEEE DataPort](https://ieee-dataport.org/documents/soil-fertility-data-fertilizer-recommendation), [Kaggle Dataset](https://www.kaggle.com/datasets/irakozekelly/fertilizer-prediction/data)).

> **Why use the Fertilizer Prediction dataset?**  
Think of `train.csv` as a patient's medical test results (vitals, lab tests, etc.), and `Fertilizer Prediction.csv` as the doctor's prescription based on those results. Without the prescription (fertilizer label), you can't train a supervised model to recommend fertilizers‚Äîeven if you know all the symptoms (features).

**üîç When is the original dataset necessary?**
- Your main dataset lacks the target label (fertilizer name/type).
- You want to train a supervised ML model to recommend fertilizers from environmental and crop conditions.
- You aim to augment or join datasets to boost model accuracy by increasing data diversity or quantity.


Our main objective is to select the best fertilizer for different weather, soil conditions and crops. per sample, using EDA, feature engineering, and multi-label ranking, evaluated by Log Loss and MAP@3 metrics.

```python
def print_unique_values(df, categorical_columns, dataset_name="Dataset"):
    print("\n" + "=" * 50)
    print(f"Unique values in {dataset_name} categorical features")
    print("=" * 50)
    for col in categorical_columns:
        unique_vals = sorted(df[col].unique())
        value_counts = df[col].value_counts()
        top_value = value_counts.index[0]
        top_freq = value_counts.iloc[0]
        print(f"{col} - Number of unique values: {len(unique_vals)}")
        print(f"Unique values: {unique_vals}")
        print(f"Top value: '{top_value}' (Frequency: {top_freq})\n")

train_cat_cols = ['Soil Type', 'Crop Type', 'Fertilizer Name']
original_cat_cols = ['Soil Type', 'Crop Type', 'Fertilizer Name']
test_cat_cols = ['Soil Type', 'Crop Type']

print_unique_values(train_data, train_cat_cols, "Train Data")
print_unique_values(original_data, original_cat_cols, "Original Data")
print_unique_values(test_data, test_cat_cols, "Test Data")
```

```python
import pandas as pd
import plotly.graph_objects as go

def data_quality_report(df_dict):
    summary_data = []

    for name, df in df_dict.items():
        missing = df.isnull().sum().sum()
        dups = df.duplicated().sum()
        total = len(df)

        summary_data.append({
            "Dataset": name,
            "Total Rows": f"{total:,}",
            "Missing Values": f"{missing:,}",
            "Duplicates": f"{dups:,}",
            "Duplicate %": f"{(dups / total):.2%}" if total > 0 else "0.00%"
        })

    summary_df = pd.DataFrame(summary_data)

    fig = go.Figure(data=[go.Table(
        header=dict(
            values=list(summary_df.columns),
            fill_color='gray',
            font=dict(color='white', size=14),
            align='center'
        ),
        cells=dict(
            values=[summary_df[col] for col in summary_df.columns],
            fill_color='lightgray',
            align='center',
            font=dict(size=13)
        )
    )])

    fig.update_layout(
        title="Data Quality Report Summary",
        margin=dict(l=20, r=20, t=40, b=20),
        height=400
    )

    fig.show()

    return summary_df
```

        <b><i>Evaluation Metric: Mean Average Precision at 3 (MAP@3)</i></b>
    </h1>
</div>

**MAP@3** is the competition‚Äôs main evaluation metric. It measures how often the correct fertilizer appears in your model‚Äôs top 3 predictions, and rewards higher ranks.

---

### What is MAP@3?

- **MAP@3** (Mean Average Precision at 3) evaluates both the presence and the rank of the correct fertilizer in your top 3 predictions.
- The earlier the correct fertilizer appears, the higher the score.

---

### How is MAP@3 Calculated?

- For each test sample, your model predicts up to **three fertilizer names** (ordered by confidence).
- If the correct fertilizer is:
  - **1st**: Score = **1.0**
  - **2nd**: Score = **0.5**
  - **3rd**: Score = **0.333...**
  - **Not in top 3**: Score = **0.0**
- **MAP@3** is the average of these scores across all test samples.

---

### Formula

$$
\text{MAP@3} = \frac{1}{U} \sum_{u=1}^{U} \sum_{k=1}^{\min(n, 3)} P(k) \times \text{rel}(k)
$$

- $U$: Number of test samples
- $P(k)$: Precision at rank $k$
- $\text{rel}(k)$: 1 if prediction at $k$ is correct, else 0

---

### Example

| Predictions          | AP@3 Score |
| -------------------- | ---------- |
| `Urea 14-35-14 DAP`  | 1.0        |
| `14-35-14 Urea DAP`  | 0.5        |
| `14-35-14 DAP Urea`  | 0.333      |
| `14-35-14 DAP 20-20` | 0.0        |

---

### Why MAP@3?

- Encourages models to rank the correct fertilizer as high as possible.
- Useful for recommendation systems where multiple options are suggested.
- In agriculture, providing the correct fertilizer within the top 3 supports better decision-making.

**Reference:** [Kaggle Evaluation Page](https://www.kaggle.com/competitions/playground-series-s5e6/overview/evaluation)
